```python
import time
import pandas as pd
import numpy as np
#create dictionary contains the 3 data files as values and names of cities as keys
CITY_DATA = {'chicago': 'chicago.csv','new york city': 'new_york_city.csv','washington': 'washington.csv'}
#The first function gets the user inputs in 3 variables (city,month,day),filter them to check validation 
#Out : city, month , day 
def user_data():
    #make user recognize the program and its benefits ask him(her)to input the data required
    print("Hello, Our Program is designed to explore data realted to Bike Share systems\n")
    print("for three major cities in the United States  Chicago, New York City, and Washington\n")
    print("you will be able to discover most important Statistics for any of the 3 cities :\n")
    city ='' #create empty variable (city)to store the input user for the city name
    while city not in CITY_DATA.keys():   #start while loop to  
        print("Chicago New York City Washington\n")
        print("please write the city name correctly :  ")
        city = input().lower()            #get the user input 
        if city not in CITY_DATA.keys():  #and check if it is in CITY_DATA dictionary as key
            print ('Incorrect City Name !! Please try again') #action if: invalid input
    print ('\n You choosed :',city) #action if : valid input
    
    #create dictionary contains the names of monthes as key and the numbers as values 
    months = {"January":1,"February":2,"March":3,"April":4,"May":5,"June":6,"All":0} 
    month ='' #create empty variable (month)to store the input user for the month name 
    print('\n Would you like to filter the Data by a specific month \n') #ask user if he(she)wants to filter  data by month
    #ask user to write month name if he want aspecific month
    #ask user to write "all" if he(she)doesn't want to filter by month
    print(' If YES Please write the month name :January, February, March, April, May, June\n' )
    print(' OR write "all" for no filter by month')
    while month not in months.keys():  #start while loop to
        month = input().title()        #get user input in variable(month),make first letter capital by (.title())
        if month not in months.keys(): #make if statement to check if (month) in (monthes) dictionary above as key 
            print ('Incorrect Month name !! Please try again') #if not found ask the user to try agin
    if month == 'All':                                        #if found:make if statement to check th value of (month) 
        print ('You didnt choose filtering by month')    # if (All) that means the user didn't choose month to filter by
    else:
        print ('You choosed: ' ,month)         # the user choosed a specific month to filter by 
    month = months[month]                    # assign the value of key=month in monthes dictionary to month 
    month = int(month)                       # convert month from srting to integer to meet (month column) specifications 
    #create list named (days) to store the dayes names  
    days= ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','All']                                                                        
    day = ''       #creat empty variable (day) to store inut user for day name          
    print('\n would you like to filter the Data by a specific Day \n')#ask user if he(she)want to filter data by day
    print('If YES please write the Day name :Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday \n')
    #ask user to write day name if he want aspecific day
    print('If NO Please write "all" ')   #ask user to write "all" if he(she)doesn't want to filter by day 
    while day not in days:    #start while loop to
        day = input().title() #get user input in variable(day),make first letter capital by (.title())
        if day not in days:   #make if statement to check if (day) in (days) list above  
            print ('Incorrect day name !! Please try again') #if not found ask the user to try agin
    if day == 'All':                 #if found:make if statement to check th value of (day) 
        print ("You didn't choose filtering by day") # if (All) that means the user didn't choose day to filter by
    else:                             # the user choosed a specific day to filter by     
        print ('You choosed :  ',day) 
    return city,month,day            # the output of the funtion (city,month,day) ready to use in another funchions


#define funtion (prepare_data-file) uses the variables(city,month,day)to Determine the data frame of the specified data
def prepare_data_file(city,month,day):
    df = pd.read_csv(CITY_DATA[city]) #load data of the choosed city into data frame
    df['Start Time']  = pd.to_datetime(df['Start Time'])#convert Start Time column to date time column
    df['month'] = df['Start Time'].dt.month #extract month  from Start Time column & creat month column
    df['day_of_week'] = df['Start Time'].dt.day_name() #extract day name from Start Time column &create day_of_week column
    df['hour'] = df['Start Time'].dt.hour #extract hour from Start Time column &create hour column
    df['start to end']=df['Start Station']+ " TO "+df['End Station'] 
     #compine Start Station column &End station column to creat start to end column 
    #make if statement to check month value                                                                  
    if month==0: 
        popular_month =df['month'].mode()[0] #calculate the mode of month column (in case of no month choosed)
        print('Most Frequent Start month:', popular_month)
    else:
        df = df[df['month'] == month]  #load data of specific month (in case of choose month ) into data frame
    #make if statement to check day value     
    if day=='All':
        popular_day=df['day_of_week'].mode()[0] #calculate the mode of day column (in case of no day choosed)
        print('most frequent day of week :' , popular_day)
    else:
        df = df[df['day_of_week'] == day]  #load data of specific day (in case of choose day) into data frame
    return df  #the output of function (df) (data frame of filterd data)

#define function (all_statistics) use df  to calculate all required statistics of the project     
def all_statistics(df):
    
    print('most frequent start hour:',df['hour'].mode()[0]) #calculate to mode of (hour) column
    print( 'Most common Start Station :',df['Start Station'].mode()[0]) #calculate the mode of (Start Station) column
    print( 'Most common End Station :',df['End Station'].mode()[0])     #calculate the mode of (End Station) column
    print('Most common trip from start to end :',df['start to end'].mode()[0]) #calculate the mode of (strat to end) column
    print('Total travel time:',df['Trip Duration'].sum()) #calculate of the sum of (Trip Duration) column
    print('Average Travel Time',df['Trip Duration'].mean()) #calculate of the mean of (Trip Duration) column
    print('User Type Counts :\n',df['User Type'].value_counts()) #display every User Type count
    try: #execute the code below unless error occurred 
        print('Gender Count :',df['Gender'].value_counts())#calculate every gender count
        print("Earliest Birth Year :yes{}\n,Most recent Birth Year:{}".format(df['Birth Year'].min(),df['Birth Year'].max()))
        #display the oldest and newest birth year for the user
    except: # if error print the sentense below
        print(' Data  of Gender, Birth Year are  not availble for Washington City' )

    
#define function (data_report) use df to display no of rows of df 
def data_report(df):
    start=0 #create variable (start) assign  value (0)to it
    end=5   #create variable (end assign  value (5)  to it
    response =input("Do you want to view sample of data rows ?\n press YES to Agree or any key to end ").lower()
    #create variable (response) to get the input user
    while response == "yes": #start while loop to repeat code below unless response != yes
        if response == "yes": #create if statement to check response value
            data = df.iloc[start:end] #create variable (data)to store the the rows from (start) to (end)
            print(data) #print the first  5 no of  rows
            response = input("do you want to continue more 5 rows ?:\n press YES to Agree or any key to end ").lower()
            # get the user input &(convert to lower case) for more print or stop
            if response == "yes": #create if statement to check the user input if (yes)
                start+=5  #increase (start) by 5
                end+=5    #increase (end) by 5           
            else:
                break  #go out of the while loop
        
             
        else:
            break #go out of the while loop
            
                 
    print("Thank you for using our project") #end of the user interaction 
#define main() function to run the above functuions in the order
def main():
    while True :    #start while loop to repeat the program unless the user wants to stop
        city,month,day = user_data()    #run (user_data) function to get (city,month,day)
        print('\n\nThe Statistics Required are :\n' )
        df = prepare_data_file(city,month,day)    #run (prepare_data_file) function to get (df)
        all_statistics(df)     #run (all_statistics(df))function to calculate all statistics of df 
        data_report(df)       #run (data_report(df)) function to display data of df
        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break
    print ("Thank you for using our Project")
    
if __name__ == '__main__':    
    main()  # call main() function

    
```
